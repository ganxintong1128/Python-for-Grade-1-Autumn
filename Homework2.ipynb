```python
import time
def  find_prime(n):
    time_start = time.time()
    number_pn = 0
    
    non_primes=set(range(4,n+1,2))
    for i in range(3, int(n**0.5)+1, 2):
        if i not in non_primes:
            for j in range(i*i, n+1, 2*i):
                non_primes.add(j)
    number_npn=len(non_primes)
    number_pn = n-1-number_npn

    time_end = time.time()    
    return time_end - time_start,  number_pn
print(find_prime(1000000))
```

```python
import time
def  find_prime(n):
    time_start = time.time()
    number_pn = 0
    
    def mod_pow(base, exp, mod):
        result = 1
        base = base % mod
        while exp > 0:
            if exp % 2 == 1:
                result = (result * base) % mod
            base = (base * base) % mod
            exp = exp // 2
        return result

    def is_prime(x):
        if x == 2 or x == 3:
            return True
        elif x % 2 == 0:
            return False
    
        u = x - 1
        t = 0
        while u % 2 == 0:
            u //= 2
            t += 1
    
        test_bases = [2, 3]
        for a in test_bases:
            if a >= x:
                continue
            y = mod_pow(a, u, x)
            if y == 1 or y == x - 1:
                continue
            for _ in range(t - 1):
                y = mod_pow(y, 2, x)
                if y == x - 1:
                    break
            else:
                return False
        return True

    for x in range(2, n + 1):
        if is_prime(x):
            number_pn += 1
    
    time_end = time.time()    
    return time_end - time_start,  number_pn
print(find_prime(1000000))
```
